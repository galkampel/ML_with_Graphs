 ## average clustering coefficient:
* clustering coefficient of node $v$:
* $N_v$ - #links between neighbors of $v$
* $k_v$- $v$'s degree
 $$
 CC(v) = \frac{N_v}{\underbrace{k_v  \choose 2}_{N_v \text{ is clique}}} = \frac{2\cdot N_v}{k_v \cdot (k_v - 1)}
 $$
* clustering coefficient of graph $G$:
$$
CC(G) = \frac{1}{|V|} \sum_{v \in V}CC(v)
$$

```python
nx.average_clustering(G)
```
