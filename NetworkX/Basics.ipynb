## Nodes
* `G.number_of_nodes()`- get total nodes in the graph
* `G.degree`- a dictionary (key=node, value=degree)
    * nx.degree(G, v) - get degree of node v in graph G
* G.neighbors(node_id)- a key iterator of all the neoghbors of node_id
### get node(s)
* G.nodes[\<node_for_adding>\]- get a single node
* G.nodes - get all nodes in the graph
* `G.nodes(data=True)`- return all nodes with their corresponding attributes
    * node[0]- node_id
    * node[1]- dictionary of attributes
### add node(s)
* add_node(node_for_adding, **attr)- add a single node to the graph
    * `node_for_adding`- a hashable object (a unique node identifier)
    * `attr`- node attributes (a dictionary)
* add_nodes_from(nodes_for_adding, **attr)- add multiple nodes to the graph
    * `nodes_for_adding`- A container of nodes (list, dict, set,...), or a container of (node, attribute dict) tuples
    * `attr`- update attributes for all nodes in nodes. 


    

```python
G.add_node(0, feature=5, label=0)

G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
  ]) #(node, attrdict)

num_nodes = G.number_of_nodes()
```

## Edges
* G.number_of_edges()- get total edges in the grpah
* G.has_edge(n1, n2)- check if an edge exists between n1 and n2

### get edge(s)
* G.edges[$(\text{node}_i, \text{node}_j)$]- get a single edge with a tuple of node identifiers
* G.edges - get all edges in the graph
* `G.edges(data=True)`- return all edges with their corresponding attributes
### add edge(s)
* add_edge(u_of_edge, v_of_edge, **attr)- add a single node to the graph
    * `u_of_edge`- a unique node identifier for node $u$
    * `v_of_edge`- a unique node identifier for node $v$
    * `attr`- edges attributes (a dictionary)
* add_nodes_from(ebunch_to_add, **attr)- add multiple nodes to the graph
    * `ebunch_to_add`- a list of 2-tuples (u, v) or 3-tuples (u, v, d) where d is a dictionary containing edge data
    * `attr`- update attributes for all nodes in nodes. 

### self loops
